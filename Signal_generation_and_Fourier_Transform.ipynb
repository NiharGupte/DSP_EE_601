## 8. Plot the following sequences for n = −7, −6, . . . , 6, 7 using stem plots. Label the plots appropriately.

## (a) δ[n]

## (b) u[n]

## (c) δ[n + 1] − δ[n − 1]

## (d) (0.5)^n u[n]

## (e) (−0.5)^n u[n]

```python
samples = 15

x = np.linspace(-7,7,samples)


#unit impulse
N = 0

y1 = np.where(x==N,1,0)

#unit step
y2 = np.where(x>=N,1,0)


N = 1
y3 = np.where(x==-N,1,0) - np.where(x==N,1,0)

y4 = np.power(0.5,x) * y2

y5 = np.power(-0.5,x) * y2

figure = plt.figure(figsize = [15,15])
row = 5
col = 2

figure.add_subplot(row,col,1)
plt.title("$\delta[n]$")
plt.xlabel("[n]")
plt.ylabel("y1[n]")
plt.stem(x,y1,linefmt = "g", basefmt = "black", markerfmt="go")

figure.add_subplot(row,col,2)
plt.title("$u[n]$")
plt.xlabel("[n]")
plt.ylabel("y2[n]")
plt.stem(x,y2,linefmt = "g", basefmt = "black", markerfmt="go")

figure.add_subplot(row,col,5)
plt.title("$\delta[n+1] - \delta[n-1]$")
plt.xlabel("[n]")
plt.ylabel("y3[n]")
plt.stem(x,y3,linefmt = "g", basefmt = "black", markerfmt="go")

figure.add_subplot(row,col,6)
plt.title("$0.5^{n}u[n]$")
plt.xlabel("[n]")
plt.ylabel("y4[n]")
plt.stem(x,y4,linefmt = "g", basefmt = "black", markerfmt="go")

figure.add_subplot(row,col,9)
plt.title("$(-0.5)^{n}u[n]$")
plt.xlabel("[n]")
plt.ylabel("y5[n]")
plt.stem(x,y5,linefmt = "g", basefmt = "black", markerfmt="go")

```

### First part about generating sine wave with following inputs :

###L : Number of samples

###A : Amplitude

###w0 : Angular frequency ( $\omega_{0}$)

###phi : Phase ($\phi$)

```python


L = int(input("Enter number of samples L : "))
A = float(input("Enter amplitude A : "))
w0 = float(input("Enter frequency w0 : "))
phi = float(input("Enter phase phi : "))

Sin = lambda x,A,L,w0,phi : A*np.sin(w0*x + phi) 

x = np.linspace(0,L-1,L)
y = Sin(x,A,L,w0,phi)

plt.stem(x,y, markerfmt = "go", linefmt = "g", basefmt = "black")
plt.title("Asin($\omega_{0} n$ + $\phi$)")
plt.xlabel("$n$")
plt.ylabel("Amplitude A")
plt.show()
```

### (a) Generate sinusoidal sequences with $\omega_{0}$ = 0.3$\pi$ and $\omega_{0}$ = 0.75$\pi$. Determine the period of each sequence from the plot and verify the result theoretically.

```python
L = 50
A = 1

x = np.linspace(0,L-1,L)
y1 = A*np.sin(0.3*x*np.pi)
y2 = A*np.sin(0.75*x*np.pi)

figure = plt.figure(figsize=[10,7])

figure.add_subplot(3,1,1)
plt.stem(x,y1, markerfmt = "go", linefmt = "g", basefmt = "black")
plt.title("sin($\omega_{0} n$)      ($\omega_{0} = 0.3 \pi$) ")
plt.xlabel("$n$")
plt.ylabel("$y1[n]$ ")

figure.add_subplot(3,1,3)
plt.stem(x,y2, markerfmt = "go", linefmt = "g", basefmt = "black")
plt.title("sin($\omega_{0} n$)      ($\omega_{0} = 0.75 \pi$) ")
plt.xlabel("$n$")
plt.ylabel("$y2[n]$ ")

plt.show()

```

Here, for the first signal, $\omega$ = 0.3$\pi$, the signal as evident from the waveform repeats itself after 20 samples, so N = 20. For the second signal, $\omega$ = 0.75$\pi$, the signal as evident from the waveform repeats itself after 8 Samples, so N = 8 here.

Theoretically, N = 2$\pi m/ \omega$ where $m$ is an integer so as to make N purely integer (period cannot be fractional in discrete domain). So, on substituting values of $\omega$ = 0.3$\pi$ and 0.75$\pi$, we get N = 20 and 8 respectively.

### (b) Generating the sequence : 

###$x[n] = sin(0.1 \pi n + 0.75 \pi) + 3 cos(0.8\pi n + 0.2\pi ) + cos(1.3\pi n)$

```python
L = 100

n = np.arange(0,L)
x = np.sin(0.1*np.pi*n + 0.75*np.pi) - 3*np.cos(0.8*np.pi*n + 0.2*np.pi) + np.cos(1.3*np.pi*n)


#For finding period, we use the list comprehension as shown below
#Since at exact periodic values, the values do not get equal, we use the absolute function.
#period here returns the list of all periods, the first element being the fundamental period 
period = [N for N in n[1:] if abs(x[N] - x[0]) < 1e-5]

plt.stem(n,x,markerfmt = "go", linefmt = "g", basefmt = "black")
plt.title("$sin(0.1 \pi  n + 0.75 \pi ) - 3cos(0.8 \pi n + 0.2 \pi) + cos(1.3 \pi n)$")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

plt.show()
print("\n Fundamental period of the given signal is : " +str(period[0]))
```

## Q 11. 

### (a) Plotting normalized DTFT for N = 5,20,100

```python

L = 150

n = np.arange(-L,L)
w = np.linspace(-3.14,3.14,512)

Window = lambda N,n : np.array([1 if i <= N and i >= -N else 0 for i in n])

DTFT = lambda w,N : np.sin((2*N+1)*w/2)/np.sin(w/2)


figure = plt.figure(figsize = [20,20])
figure.add_subplot(11,2,1)
plt.stem(n,Window(5,n))
plt.title("DT rectangular window for N = 5")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,2)

plt.title("Scaled DTFT for N = 5")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,DTFT(w,5)/ np.max(DTFT(w,5)))

figure.add_subplot(11,2,5)
plt.stem(n,Window(20,n))
plt.title("DT rectangular window for N = 20")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,6)
plt.title("Scaled DTFT for N = 20")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,DTFT(w,20)/np.max(DTFT(w,20)))


figure.add_subplot(11,2,9)
plt.stem(n,Window(100,n))
plt.title("DT rectangular window for N = 100")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,10)
plt.title("Scaled DTFT for N = 100")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,DTFT(w,100)/np.max(DTFT(w,100)))





plt.show()
```

As window size increases, the sinc pulse of DTFT gets compressed

#### Magnitude and Phase plot of DTFT with Rectangular Window function N = 5

```python
N = 5
DTFT_scaled = DTFT(w,N)/np.max(DTFT(w,N))

#Magnitude and Phase Plot
Magnitude = np.absolute(DTFT_scaled)
Phase = np.angle(DTFT_scaled)

figure = plt.figure(figsize=[20,15])

figure.add_subplot(2,2,1)
plt.title("Magnitude part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$|(X(\omega)|$")
plt.axhline(color = "black")
plt.plot(w,Magnitude)

figure.add_subplot(2,2,2)
plt.title("Phase part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$angle (X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,Phase)
```

#### Imaginary and Real Components

```python
#Imaginary and Real components

figure = plt.figure(figsize = [20,15])

Real = np.real(DTFT_scaled)
Imaginary = np.imag(DTFT_scaled)

figure.add_subplot(2,2,1)
plt.title("Real part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$Re(X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,Real)

figure.add_subplot(2,2,2)
plt.title("Imaginary part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$Im(X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,Imaginary)


```

### (b) Triangular window, applying all same functions as above

```python
L = 150

n = np.arange(-L,L)
w = np.linspace(-3.14,3.14,512)

Window_triangular = lambda N,n : np.array([(1 - abs(i)/N) if abs(i) <= N else 0 for i in n])

DTFT_triangular = lambda w,N : np.sin((N)*w/2)**2/(N*np.sin(w/2)**2)

figure = plt.figure(figsize = [20,30])
figure.add_subplot(11,2,1)
plt.stem(n,Window_triangular(5,n))
plt.title("DT triangular window for N = 5")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,2)

plt.title("Scaled DTFT for N = 5")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,DTFT_triangular(w,5)/ np.max(DTFT_triangular(w,5)))

figure.add_subplot(11,2,5)
plt.stem(n,Window_triangular(20,n))
plt.title("DT rectangular window for N = 20")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,6)
plt.title("DTFT for N = 20")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.plot(w,DTFT_triangular(w,20)/np.max(DTFT_triangular(w,20)))


figure.add_subplot(11,2,9)
plt.stem(n,Window_triangular(100,n))
plt.title("DT rectangular window for N = 100")
plt.xlabel("$n$")
plt.ylabel("$x[n]$")

figure.add_subplot(11,2,10)
plt.title("DTFT for N = 100")
plt.xlabel("$\omega$")
plt.ylabel("$X(\omega)$")
plt.plot(w,DTFT_triangular(w,100)/np.max(DTFT_triangular(w,100)))





plt.show()

```

Again, as the length of triangular function increases, the normalized DTFT gets compressed

#### Magnitude and Phase spectrum for Triangular window

```python
N = 5
DTFT_scaled = DTFT_triangular(w,N)/np.max(DTFT_triangular(w,N))

#Magnitude and Phase Plot
Magnitude = np.absolute(DTFT_scaled)
Phase = np.angle(DTFT_scaled)

figure = plt.figure(figsize=[20,15])

figure.add_subplot(2,2,1)
plt.title("Magnitude part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$|(X(\omega)|$")
plt.axhline(color = "black")
plt.plot(w,Magnitude)

figure.add_subplot(2,2,2)
plt.title("Phase part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$angle (X(\omega)$")
plt.plot(w,Phase)
```

#### Real and Imaginary components

```python
#Imaginary and Real components

figure = plt.figure(figsize = [20,15])

Real = np.real(DTFT_scaled)
Imaginary = np.imag(DTFT_scaled)

figure.add_subplot(2,2,1)
plt.title("Real part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$Re(X(\omega)$")
plt.axhline(color = "black")
plt.plot(w,Real)

figure.add_subplot(2,2,2)
plt.title("Imaginary part of DTFT")
plt.xlabel("$\omega$")
plt.ylabel("$Im(X(\omega)$")
plt.plot(w,Imaginary)
```

Solution: 

The difference between first the DTFT of both the function is the phase spectrum. Since DTFT of rectangular function has sine components in numerator and denominator with odd power, the phase varies between $\pi$ and 0.

On the other hand, DTFT of triangular function has sine components in numerator and denominator with even power, the phase does not vary between $\pi$ and 0. But remains stable at 0.

Also for same $N$, the magnitude becomes zero at different values for both the functions as seen in the below plot
